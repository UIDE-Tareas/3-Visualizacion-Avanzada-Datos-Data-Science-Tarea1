<h2>Visualización de datos con Matplotlib</h2>

[Matplotlib](https://matplotlib.org/stable/index.html) es una librería en python para crear visualizaciones estáticas e interactivas.

<img src="./images/figure.png" width="55%">

En matplotlib existen dos tipos de formas de definir y construir visualizaciones:
- **Explicita:** *OOS* - *object-oriented style*. Se crean la figura y los ejes, sobre los cuales se empiezan a llamar los métodos para graficar.
- **Implicita:** Se utilizan funciones de `pyplot` para graficar. Este se encarga de gestionar el *OOS* para generar las visualizaciones.

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

Ejemplo de la forma **explicita**, donde definimos un `subplot` el cual retorna una figura `fig` y unos ejes `ax`. Note que logramos graficar haciendo uso de los llamados a los ejes `ax`.

```python
x = np.linspace(0, 2, 100)
print(x)
```

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots(figsize=(8, 4), layout='constrained')
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label') 
ax.set_title("Simple Plot using Object-oriented style")
ax.legend();
```

En la forma **implicita** no definimos una figura o ejes. Lo que hacemos es llamar funciones de `pyplot`, y este se va encargando de construir nuestra gráfica.

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(8, 4))
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot using pyplot-style")
plt.legend();
```

<h3>Tipos de gráficos</h3>

[Matplotlib](https://matplotlib.org/stable/index.html) ofrece un gran número de [representaciones gráficas](https://matplotlib.org/stable/plot_types/index.html#plot-types). Las cuales mediante la llamada a la función de cada gráfico nos retorna (mediate el uso implícito) el gráfico.

- `plt.plot(**args)`: Nos permite graficar lineas, relaciones en 2D
- `plt.scatter(**args)`: Graficos de dispersión
- `plt.bar(**args)`: Graficos de barras
- `plt.hist(**args)`: Nos permite graficar distribuciones
- `plt.boxplot(**args)`: Para representar diagramas de cajas
- `plt.violinplot(**args)`: Para representar diagramas de violín
- `plt.pie(**args)`: Para representar proporciones

<h2>Subplots en matplotlib</h2>

Los [subplots o mosaicos](https://matplotlib.org/stable/users/explain/axes/mosaic.html) en matplotlib nos permiten agrupar gráficos. Estos tienen características diferentes:

- [`subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.subplots.html#matplotlib.figure.Figure.subplots): Permiten generar *layouts* para agregar varios *plots* a una única figura.
    - `nrows`: número de filas de la cuadricula
    - `ncols`: número de columnas de la cuadricula

    Donde cada posición $(i,j)$ representa una gráfica.
    
- [`subplot_mosaic`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.subplot_mosaic.html#matplotlib.figure.Figure.subplot_mosaic): Permite generar figuras en un espacio de cuadricula no uniforme. Por lo cual podemos combinar filas y/o columnas.
    - `mosaic`: Lista de listas con la distribución del mosaico.

Profundicemos en los [`subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.subplots.html#matplotlib.figure.Figure.subplots):

```python
# Nos permite definir el subplot, dentro del tamaño pre-definido de la figura
plt.subplots(2,3, layout='constrained', figsize=(10,5));
```

```python
plt.figure(figsize=(10,5), layout="constrained")

plt.subplot(2,2,1, facecolor="#FFEEEB")
plt.scatter(df.numero_habitaciones, df.precio, color="green")
plt.xlabel("Habitaciones")
plt.ylabel('Precio')
plt.title("habitaciones vs precio")

plt.subplot(2,2,2, facecolor="#F7FFF7")
plt.scatter(df.metro_cuadrado, df.precio, color="red")
plt.xlabel("$m^2$")
plt.ylabel('Precio')
plt.title("Metro cuadrado vs precio")

plt.subplot(2,2,3, facecolor="#FFFFED")
plt.scatter(df.metro_cuadrado,
            df.numero_habitaciones,
           color="black")
plt.xlabel("Metro cuadrado")
plt.ylabel('Habitaciones')
plt.title("Metro cuadrado vs habitaciones");
```

Las limitaciones de los subplot se deben a que siempre se construye una cuadricula uniforme de $n$ filas por $m$ columnas, por lo cual, se genera un espacio de gráficas de $n*m$. Si no se pintan las $n*m$ gráficas, quedarán espacios vacios. A su vez, todas las gráficas tendrán el mismo tamaño, por lo cual no podemos resaltar gráficas que consideremos de mayor impacto.

Exploremos los [`subplot_mosaic`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.subplot_mosaic.html#matplotlib.figure.Figure.subplot_mosaic):

Nota: Para esto, debemos entender primero el concepto de mosaico:

<img src="./images/mosaics.png" width="90%">

```python
m = [["A", "C"],
     ["A", "D"]]

fig, axes = plt.subplot_mosaic(mosaic=m,
                               layout="constrained",
                               figsize=(7,6))
axes["A"].set_title("Metro cuadrado vs precio")
axes["A"].scatter(df.metro_cuadrado, df.precio, color="red")
axes["A"].set_xlabel("precio")
axes["A"].set_ylabel("$m^2$")

axes["C"].set_title("Habitaciones vs precio")
axes["C"].scatter(df.numero_habitaciones, df.precio, color="green")
axes["C"].set_xlabel("numero habitaciones")
axes["C"].set_ylabel("precio")

axes["D"].set_title("Metro cuadrado vs habitaciones")
axes["D"].scatter(df.metro_cuadrado, df.numero_habitaciones, color="black")
axes["D"].set_xlabel("$m^2$")
axes["D"].set_ylabel("habitaciones")

plt.show()
```

Con los mosaicos podemos combinar filas y columnas para brindar mayor atención a gráficas en especiífico.
